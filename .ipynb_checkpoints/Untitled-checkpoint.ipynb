```python
import cv2 as cv
import pandas as pd
import numpy as np
import textract
import matplotlib.pyplot as plt
```

```python
image = cv.imread("pan.jpg")
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(image)
```

```python
img_threshold = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY, 11, 2)
plt.imshow(img_threshold, cmap = 'gray')
```

```python
kernel = np.ones((1, 1), np.uint8)
img_erosion = cv.erode(img_threshold, kernel, iterations=1)
img_dilation = cv.erode(img_erosion, kernel, iterations=1)

plt.imshow(img_dilation, cmap = 'gray')
```

```python
kernel = np.ones((2, 2), np.uint8)


morphology_ex = cv.morphologyEx(img_threshold, cv.MORPH_GRADIENT, kernel)
plt.imshow(morphology_ex, cmap = 'gray')
```

```python
plt.imsave("morphology_ex.png", morphology_ex)
plt.imsave("threshold.png", img_threshold)
```

```python
text = textract.process('threshold.png')

text
```
